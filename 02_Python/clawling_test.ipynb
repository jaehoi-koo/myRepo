```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.webdriver import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import csv
import time
from webdriver_manager.chrome import ChromeDriverManager

# 1ï¸âƒ£ Selenium ì„¤ì • (ë¸Œë¼ìš°ì € ìë™ ì‹¤í–‰)
options = Options()
options.add_argument("--headless")  # ë¸Œë¼ìš°ì € ì°½ ì—†ì´ ì‹¤í–‰ (ì›í•˜ë©´ ì£¼ì„ ì²˜ë¦¬)
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# 2ï¸âƒ£ ìë™ì°¨ ë¦¬ì½œ ì‚¬ì´íŠ¸ ì ‘ì†
url = "https://www.car.go.kr/ri/stat/list.do"
driver.get(url)

# í˜ì´ì§€ ë¡œë“œ ëŒ€ê¸° (WebDriverWait ì‚¬ìš©)
wait = WebDriverWait(driver, 10)

# 3ï¸âƒ£ CSV íŒŒì¼ ì¤€ë¹„
filename = "recall_data.csv"
with open(filename, "w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["ë¦¬ì½œë²ˆí˜¸", "ì œì¡°ì‚¬", "ì°¨ëª…", "ë¦¬ì½œì‚¬ìœ ", "ë¦¬ì½œì¼ì"])  # ì»¬ëŸ¼ í—¤ë” ì¶”ê°€

    while True:
        try:
            # 4ï¸âƒ£ í…Œì´ë¸” ë°ì´í„° ê°€ì ¸ì˜¤ê¸°
            wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "table tbody tr")))  # í…Œì´ë¸”ì´ ë¡œë“œë  ë•Œê¹Œì§€ ëŒ€ê¸°
            rows = driver.find_elements(By.CSS_SELECTOR, "table tbody tr")

            for row in rows:
                cols = row.find_elements(By.TAG_NAME, "td")
                if len(cols) >= 5:
                    data = [col.text.strip() for col in cols[:5]]  # ì• 5ê°œ ë°ì´í„° ì €ì¥
                    writer.writerow(data)
                    print(data)  # í„°ë¯¸ë„ ì¶œë ¥ (ë””ë²„ê¹…ìš©)

            # 5ï¸âƒ£ ë‹¤ìŒ í˜ì´ì§€ ë²„íŠ¼ í´ë¦­ (XPath ìˆ˜ì •)
            next_button = wait.until(EC.element_to_be_clickable((By.XPATH, "//a[contains(text(), 'ë‹¤ìŒ')]")))

            if "disabled" in next_button.get_attribute("class"):  # ë²„íŠ¼ì´ ë¹„í™œì„±í™”ë˜ë©´ ì¢…ë£Œ
                print("âœ… ëª¨ë“  í˜ì´ì§€ í¬ë¡¤ë§ ì™„ë£Œ!")
                break

            next_button.click()
            time.sleep(3)  # í˜ì´ì§€ ë¡œë”© ëŒ€ê¸°
        except Exception as e:
            print("âŒ ë‹¤ìŒ í˜ì´ì§€ ë²„íŠ¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŒ:", e)
            break

# 6ï¸âƒ£ í¬ë¡¤ë§ ì¢…ë£Œ
driver.quit()
print(f"ğŸ‰ ë°ì´í„°ê°€ '{filename}' íŒŒì¼ì— ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
```
